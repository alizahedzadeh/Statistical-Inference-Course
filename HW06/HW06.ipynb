𝑥3= dollar advertising expenditure in the territory 
𝑥4= weighted average of the company’s market share in the territory for the previous 
four years 
𝑥5 = change in the company’s market share in the territory over the previous four years A random sample of 26 observations shows the following results:
</span>





#### <span style='color:#E74C3C; font-family: "Arial", sans-serif; font-size: 1.1em; font-weight: bold;'>Answer</span>

##### <span style='color:#3498DB; font-family: "Helvetica", Arial, sans-serif; font-size: 1.2em; font-weight: bold;'>Part a:</span>

<span style='font-family: "Segoe UI", Tahoma, Geneva, Verdana, sans-serif; font-size: 1.2em; font-weight: bold;'>
    Test the overall significance (i.e., validity) of the multiple regression model 
using a 5% significance level.
</span>




**Hypotheses**

- Null Hypothesis $H_0$: All regression coefficients $\beta_1, \beta_2, \beta_3, \beta_4, \beta_5$ are equal to zero. This means the model has no explanatory power.
- Alternative Hypothesis $H_1$: At least one regression coefficient is not equal to zero. This means the model has some explanatory power.

**F-Test Calculation**

Given the information from the partial ANOVA table:

- SSR (Sum of Squares for Regression) = 39,500
- SSE (Sum of Squares for Error) = 3,500
- SST (Total Sum of Squares) = 43,000
- Number of observations $n$ = 26
- Number of predictors $k$ = 5

We calculate the Mean Square for Regression (MSR) and Mean Square for Error (MSE):

$$
\text{MSR} = \frac{\text{SSR}}{k} = \frac{39500}{5} = 7900
$$

$$
\text{MSE} = \frac{\text{SSE}}{n - k - 1} = \frac{3500}{26 - 5 - 1} = \frac{3500}{20} = 175
$$

The F-statistic is calculated as:

$$
F = \frac{\text{MSR}}{\text{MSE}} = \frac{7900}{175} \approx 45.14
$$

**Determine the Critical Value**

At a 5% significance level $\alpha = 0.05$, we need the critical value of the F-distribution with $k$ (5) and $n - k - 1$ (20) degrees of freedom. We look up the critical value $F_{\alpha, k, n-k-1}$ in an F-table or use statistical software.

**The critical value for $F_{0.05, 5, 20}$ is approximately 2.71.**



Since 45.14 > 2.71, we reject the null hypothesis.

**t-Test Calculation**

The t-value for each coefficient is calculated as:

$$
t_i = \frac{\text{Coefficient} \ (\beta_i)}{\text{Standard Error}}
$$

Given the coefficients and their standard errors:

The degrees of freedom for the t-tests is $n - k - 1 = 26 - 5 - 1 = 20$.

**Calculate t-values**

$$

**Calculating P-values**

For a two-tailed test at a 5% significance level $\alpha = 0.05$ with 20 degrees of freedom.

```r
# Given t-values
t_values <- c(-2.65, 3.00, 4.00, 3.25, 1.47, 1.62)

# Degrees of freedom
df <- 20

# Calculate the p-values
p_values <- 2 * pt(-abs(t_values), df)

# Display the p-values
p_values
```

- Degrees of freedom $df$: $n - k - 1 = 26 - 5 - 1 = 20$

**Find the critical t-value**

For a 98% confidence interval, the significance level $\alpha = 0.02$, and for a two-tailed test, $\alpha/2 = 0.01$.

Using a t-table, we find the critical t-value for $\alpha/2 = 0.01$ with 20 degrees of freedom. The critical t-value $t_{0.01, 20}$ is approximately 2.528.

**Calculate the margin of error**

The margin of error (ME) is calculated as:
$$
\text{ME} = t_{\alpha/2, \text{df}} \times \text{SE}(\beta_0)
$$

$$
\text{ME} = 2.528 \times 420.00 \approx 1061.76
$$

The confidence interval (CI) is calculated as:
$$
\text{CI} = \beta_0 \pm \text{ME}
$$

$$
\text{CI} = -1113.00 \pm 1061.76
$$

Thus, the lower and upper bounds of the confidence interval are:
$$
\text{Lower bound} = -1113.00 - 1061.76 = -2174.76
$$
$$
\text{Upper bound} = -1113.00 + 1061.76 = -51.24

The 98% confidence interval for the intercept is approximately (-2174.76, -51.24).

    If a simple regression analysis is performed with 𝑥1 as the only independent 
variable, the Sum of Squares for Error (SSE) is three times as large as the SSE in the 
above multiple regression model with 5 independent variables. Test if this simple 
regression model is significant or not.
</span>




**Given:**

- Sum of Squares for Error (SSE) in the multiple regression model with 5 independent variables is 3,500.
- SSE in the simple regression model with $x_1$ as the only independent variable is three times the SSE of the multiple regression model: $3 \times 3500 = 10,500$.

We need to test the significance of the simple regression model.

**Hypotheses**
- Null Hypothesis $H_0$: The simple regression model is not significant $\beta_1 = 0$.
- Alternative Hypothesis $H_1$: The simple regression model is significant $\beta_1 \neq 0$.


**Total Sum of Squares (SST)**:
   - From the given data, SST = 43,000 (same as in the multiple regression model).

**Sum of Squares for Regression (SSR)** for the simple regression model:
   $$
   \text{SSR} = \text{SST} - \text{SSE} = 43,000 - 10,500 = 32,500
   $$

**Degrees of Freedom**:
   - For SSR: $df_{SSR} = 1$
   - For SSE: $df_{SSE} = n - 2 = 26 - 2 = 24$
   - Total degrees of freedom: $df_{Total} = n - 1 = 26 - 1 = 25$

**Mean Squares**:
   $$
   \text{MSR} = \frac{\text{SSR}}{df_{SSR}} = \frac{32,500}{1} = 32,500
   $$
   $$
   \text{MSE} = \frac{\text{SSE}}{df_{SSE}} = \frac{10,500}{24} \approx 437.5
   $$

**F-statistic**:
   $$
   F = \frac{\text{MSR}}{\text{MSE}} = \frac{32,500}{437.5} \approx 74.29
   $$

**Determine the critical value**:
   - For a 5% significance level $\alpha = 0.05$, with $df_{SSR} = 1$ and $df_{SSE} = 24$, we look up the critical value from the F-distribution table.
   - The critical value $F_{0.05, 1, 24}$ is approximately 4.26.

**Compare the F-statistic to the critical value**:
   - Calculated F-statistic: 74.29
   - Critical value: 4.26

Since 74.29 > 4.26, we reject the null hypothesis.

The simple regression model with $x_1$ as the only independent variable is significant at the 5% significance level. This means that $x_1$ significantly predicts the dependent variable (sales performance) on its own.

```r
# Load the necessary libraries
library(ggplot2)
library(mgcv) # for generalized additive models (GAMs)

# Fit a GAM with smoothing splines for both age and lpsa
gam_fit <- gam(lweight ~ s(age) + s(lpsa), data = prostate)

# Summary of the model
summary(gam_fit)

# Predict values over a grid for age
age_grid <- seq(min(prostate$age), max(prostate$age), length.out = 100)
lpsa_constant <- mean(prostate$lpsa)
predicted_age <- predict(gam_fit, newdata = data.frame(age = age_grid, lpsa = lpsa_constant))

# Predict values over a grid for lpsa
lpsa_grid <- seq(min(prostate$lpsa), max(prostate$lpsa), length.out = 100)
age_constant <- mean(prostate$age)
predicted_lpsa <- predict(gam_fit, newdata = data.frame(age = age_constant, lpsa = lpsa_grid))

# Plot the effect of age
plot(prostate$age, prostate$lweight, 
     xlab = "Age", ylab = "Log Weight (lweight)", 
     main = "Log Weight vs Age with Smoothing Spline Fit",
     col = rgb(0, 0, 1, 0.5), pch = 16)
lines(age_grid, predicted_age, col = "red", lwd = 2)
legend("topright", legend = c("Data", "Smoothing Spline Fit for Age"), 
       col = c(rgb(0, 0, 1, 0.5), "red"), pch = c(16, NA), lwd = c(NA, 2))

# Plot the effect of lpsa
plot(prostate$lpsa, prostate$lweight, 
     xlab = "Log PSA (lpsa)", ylab = "Log Weight (lweight)", 
     main = "Log Weight vs Log PSA (lpsa) with Smoothing Spline Fit",
     col = rgb(0, 0, 1, 0.5), pch = 16)
lines(lpsa_grid, predicted_lpsa, col = "red", lwd = 2)
legend("topright", legend = c("Data", "Smoothing Spline Fit for Log PSA"), 
       col = c(rgb(0, 0, 1, 0.5), "red"), pch = c(16, NA), lwd = c(NA, 2))
```

```r
# Load the necessary libraries
library(ggplot2)
library(mgcv)

# Fit a GAM with smoothing splines for both age and lpsa
gam_fit <- gam(lweight ~ s(age) + s(lpsa), data = prostate)

# Calculate the residuals
residuals <- resid(gam_fit)

# Create a data frame for plotting
residuals_data <- data.frame(Age = prostate$age, LPSA = prostate$lpsa, Residuals = residuals)

# Plot the residuals vs Age
ggplot(residuals_data, aes(x = Age, y = Residuals)) +
  geom_point(color = rgb(0, 0, 1, 0.5)) +
  geom_hline(yintercept = 0, linetype = "dashed", color = "red") +
  labs(title = "Residuals vs Age", x = "Age", y = "Residuals") +
  theme_minimal()

# Plot the residuals vs LPSA
ggplot(residuals_data, aes(x = LPSA, y = Residuals)) +
  geom_point(color = rgb(0, 0, 1, 0.5)) +
  geom_hline(yintercept = 0, linetype = "dashed", color = "red") +
  labs(title = "Residuals vs LPSA", x = "Log PSA (LPSA)", y = "Residuals") +
  theme_minimal()
```
